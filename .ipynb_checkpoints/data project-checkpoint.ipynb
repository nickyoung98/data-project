```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("country_vaccinations.csv")
```

The dataset portrays the covid 19 vaccination progress of every country
Comparing Australia's vaccination progress to 8 other countries:

Australia
Canada
China
France
Germany
United States
United Kingdom
New Zealand
Russia

```python
countries =  ["Australia", "Canada", "China", "France", "Germany", "United States", "United Kingdom", "New Zealand", "Russia" ]
def get_specific_countries(dataframe, countries):
    return [dataframe.loc[dataframe["country"] == country] for country in countries]

country_dataframes = get_specific_countries(df, countries)

            
```

```python
def max_bar_plot(countries_df, var,title):
    #add title, xlabel, ylabel
    maxes = [country[var].max() for country in countries_df]
    plt.figure(figsize=(15,8))
    plt.bar(countries, maxes, color = ["red", "blue", "green", "teal", "purple", "yellow", "orange", "pink", "black"])
    plt.title(title)
    plt.xlabel("Countries")
    plt.ylabel("Vaccination %")
    plt.show()

```

How much of Australia and the 8 countries population is vaccinated?

```python
max_bar_plot(country_dataframes, "people_fully_vaccinated_per_hundred", "% of Population Fully Vaccinated")
```

What is the daily vaccinations per million?

```python
def basic_line_graph(country_dataframes, variable):
    plt.figure(figsize=(50,20))
    dates = []
    for i in range(len(country_dataframes)):
        dates.append((i, country_dataframes[i]["date"].iloc[0]))
    dates = sorted(dates, key=lambda x:x[1])
    for country in dates:
        plt.plot(country_dataframes[country[0]]["date"], country_dataframes[country[0]][variable])
    legend = []
    for tuple in dates:
        legend.append(country_dataframes[tuple[0]]["country"].iloc[0])
    plt.legend(legend,fontsize=20, borderpad=6)
    plt.title(f"{variable} for multiple countries", fontsize=30)
    plt.xlabel("Date", fontsize=30)
    plt.ylabel("Daily Vaccinations", fontsize=30)
    plt.xticks(rotation=90)
    plt.yticks(fontsize=20)
    plt.show()


```

```python
basic_line_graph(country_dataframes[0:8], "daily_vaccinations_per_million")

```

```python

basic_line_graph(country_dataframes[0:1], "daily_vaccinations_per_million")
```

I wanted to find out how Australia was progressing with daily vaccinations compared to the day before

```python
aus = country_dataframes[0]
aus = aus.dropna()
aus["growth"] = np.log(aus["daily_vaccinations_per_million"]).diff()
aus = aus.dropna()
plt.figure(figsize=(20,5))
plt.plot(aus["date"], aus["growth"])
plt.plot(aus["date"], [0.0 for i in range(len(aus))])
plt.xlabel("Dates")
plt.ylabel("Vaccinations Per Million %")
plt.title("Australia's Daily Vaccinations Compared To Previous Day")
plt.xticks(rotation=90)
plt.show()
```

Comparing the population of Australians getting one dose and two doses (fully vaccinated)  

```python
plt.figure(figsize=(20,5))
plt.plot(aus["date"], aus["people_vaccinated_per_hundred"])
plt.plot(aus["date"], aus["people_fully_vaccinated_per_hundred"])
plt.legend(["first dose", "second dose"])
plt.title("First and Second Dosage % of Australia")
plt.xlabel("Dates")
plt.ylabel("% of Population")
plt.xticks(rotation=65)
plt.show()
```
